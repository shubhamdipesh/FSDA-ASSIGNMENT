```python
#Ans.3def fib(n):
    a = 0
    b = 1
    if n == 1:
        print(a)
    else:
        print(a)
        print(b)
        for i in range(2,n):
            c = a + b
            a = b
            b = c
            print(c)
        fib(10)
```

```python
#Ans.4 lower = int(input("Enter lower range:"))
upper = int(input("Enter upper range:"))
for num in range(lower ,upper+1):
    sum = 0
    temp = num
    while temp>0:
        digit = temp%10
        sum+= digit**3
        temp//= 10
        if num == sum:
            print(num)

```
